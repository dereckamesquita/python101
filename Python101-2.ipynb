### Bucle While

Ahora, imagina que quieres jugar a videojuegos hasta que sea hora de dormir, pero no sabes cuánto tiempo será.
En Python, usas un bucle while para hacer esto. Le dices a la computadora: "¡Sigue jugando hasta que te diga que paremos!".
Aquí hay un ejemplo:

```python
 # Inicializamos la variable número con un valor que no sea 5
numero = 0

while numero != 5:
    numero = numero + 1
    print("Número nuevo:", numero)

print("¡Encontraste el número 5!")
```

```python
import random

contar = 0
numero = 0
while numero != 5:
    numero = random.randint(1, 10)  # Genera un número aleatorio entre 1 y 10
    print("Número generado:", numero)
    contar = contar + 1

print("¡Encontraste el número 5! Despues de: ", contar)
```

```python
import random

while True: # Nos permite detenerlo con un break
    numero = random.randint(1, 10)  # Genera un número aleatorio entre 1 y 10

    print("Número generado:", numero)

    if numero == 5:
        print("¡Encontraste el número 5!")
        break  # Sale del bucle cuando se encuentra el 5
    else:
      print("nada")
```

### Agrupamiento

```python
import pandas as pd

data = {
    'Producto': ['A', 'B', 'A', 'B', 'A', 'C'],
    'Ventas': [100, 150, 200, 120, 180, 90],
    'Region': ['Norte', 'Sur', 'Norte', 'Sur', 'Este', 'Oeste']
}

df = pd.DataFrame(data)
```

Aplicación de la Función groupby():

Utilizamos la función groupby() y pasamos el nombre de la columna (o columnas) por la cual deseamos agrupar los datos. Por ejemplo, para agrupar por 'Producto':

```python
grupos_por_producto = df.groupby('Producto')
grupos_por_producto
```

```python
suma_ventas_por_producto = grupos_por_producto['Ventas'].sum()
suma_ventas_por_producto
```

```python
# Agrupar los datos por las columnas 'Producto' y 'Region' y calcular la suma de ventas para cada grupo
ventas_por_producto_region = df.groupby(['Producto', 'Region'])['Ventas'].sum()

# Mostrar los resultados
print(ventas_por_producto_region)
```

```python
# Agrupar los datos solo por la columna 'Region' y calcular tanto el promedio como la suma de ventas para cada región
ventas_por_region = df.groupby('Region')['Ventas'].agg(['mean', 'sum'])

# Mostrar los resultados
print(ventas_por_region)
```

## Finanzas: CAMP ?

$ (R_i - rf) = \beta (R_m - rf) $

Exceso de Retorno: $ R_i - rf $

Exceso de Rendimiento de mercado : ($R_m - rf$)


Modelo económetrico:

$Y = \beta_0 + \beta_1 X$

Donde Y, será igual al Exceso de rendimiento de la acción, contra el exceso de rendimiento del mercado.

```python
import pandas as pd
import numpy as np
import yfinance as yf
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Lista de tickers de acciones para análisis (puedes cambiar estos ejemplos)
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA', 'JPM', 'GOOG', 'BRK-A', 'V',
           'JNJ', 'PG', 'NVDA', 'PYPL', 'UNH', 'BAC', 'MA', 'HD', 'DIS', 'INTC',
           'CMCSA', 'VZ', 'CSCO', 'KO', 'PFE', 'T', 'WMT', 'PEP', 'NFLX', 'ABT']

# Fechas de inicio y fin para los datos históricos (2 años)
start_date = '2021-09-28'
end_date = '2023-09-28'

# Crear un DataFrame vacío para almacenar los resultados de las regresiones
results = []

for ticker in tickers:
    # Descargar datos de precios de acciones
    stock_data = yf.download(ticker, start=start_date, end=end_date)

    # Descargar datos de factores de riesgo (por ejemplo, rendimiento del mercado)
    market_data = yf.download('^GSPC', start=start_date, end=end_date)

    # Calcular los rendimientos diarios de las acciones y los factores de riesgo
    stock_returns = stock_data['Adj Close'].pct_change().dropna()
    market_returns = market_data['Adj Close'].pct_change().dropna()

    # Calcular el Exceso de Retorno de la Acción (Y) y el Exceso de Rendimiento del Mercado (X)
    risk_free_rate = 0.03  # Cambia esta tasa según tus necesidades
    Y = stock_returns - risk_free_rate
    X = market_returns - risk_free_rate

    # Establecer un modelo de regresión lineal simple
    X = sm.add_constant(X)  # Variable independiente con término constante
    model = sm.OLS(Y, X).fit()

    # Almacenar los resultados de la regresión en la lista de resultados
    results.append({
        'Ticker': ticker,
        'Beta': model.params['Adj Close'],
        'Exceso_Retorno_Accion': Y.mean(),
        'Exceso_Rendimiento_Mercado': X.mean()
    })

# Crear un DataFrame con los resultados de todas las regresiones
results_df = pd.DataFrame(results)

# Plotear Beta vs. Exceso de Rendimiento de la Acción
plt.figure(figsize=(10, 6))
plt.scatter(results_df['Beta'], results_df['Exceso_Retorno_Accion'], alpha=0.5)
plt.title('Beta vs. Exceso de Rendimiento de la Acción')
plt.xlabel('Beta')
plt.ylabel('Exceso de Rendimiento de la Acción')
plt.grid(True)
plt.show()
```

```python
# Variables para la regresión
X = results_df['Beta']  # Variable independiente (Beta)
Y = results_df['Exceso_Retorno_Accion']  # Variable dependiente (Exceso de Retorno)

# Calcular los coeficientes de la regresión lineal (pendiente y término constante)
slope, intercept = np.polyfit(X, Y, 1)

# Crear una función de regresión lineal a partir de los coeficientes
regression_line = np.poly1d([slope, intercept])

# Crear un rango de valores de Beta para la línea de regresión
x_range = np.linspace(X.min(), X.max(), 100)

# Calcular los valores de Exceso de Retorno correspondientes a la línea de regresión
y_range = regression_line(x_range)

# Crear el gráfico de dispersión y la línea de regresión
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, alpha=0.5)
plt.plot(x_range, y_range, color='red', linestyle='--', linewidth=2, label=f'Regresión Lineal (y = {slope:.4f}x + {intercept:.4f})')
plt.title('Beta vs. Exceso de Rendimiento de la Acción')
plt.xlabel('Beta')
plt.ylabel('Exceso de Rendimiento de la Acción')
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()

```

```python
X = results_df['Beta']  # Variable independiente (Beta)
Y = results_df['Exceso_Retorno_Accion']  # Variable dependiente (Exceso de Retorno)

# Agregar una constante a la variable independiente (intercepto)
X = sm.add_constant(X)

# Realizar la regresión lineal
model = sm.OLS(Y, X).fit()

# Mostrar los resultados de la regresión
print(model.summary())
```

## Econometría: Errores no normales

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generar datos aleatorios
X = np.random.rand(100, 1)
error =  np.random.randn(100, 1) # Variable independiente (100 puntos)
Y = 4 + 2 * X + error  # Variable dependiente con ruido

# Crear y ajustar el modelo de regresión lineal en una línea
modelo = LinearRegression().fit(X, Y)

# Graficar los datos y la línea de regresión en una línea
plt.scatter(X, Y)
plt.plot(X, modelo.predict(X), color='red')
plt.show()

# Imprimir los coeficientes de la regresión
print(f'Pendiente: {modelo.coef_[0][0]:.4f}')
print(f'Intercepto: {modelo.intercept_[0]:.4f}')
```

```python
# Realizar la regresión lineal
X = sm.add_constant(X)

model = sm.OLS(Y, X).fit()

# Mostrar los resultados de la regresión
print(model.summary())
```
